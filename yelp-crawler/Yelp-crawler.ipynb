```python
import pandas as pd
data = pd.read_csv('cities-ca.csv')
data.head(5)
```

```python
from yelpapi import YelpAPI
import json
from slugify import slugify

_KEY = "hifRyVk-8J_S5VDz48d4V1Pq2DE3gNPvlCT37V_UZzrNh_6XvAHlTziSI5w6JzoJkELP85VR7xGLnKHhpP4hyO5FESeUDRVu3hZX2h8k2d7fFaH73_XdaQ3PxNvdXHYx" #  Replace this with your real API key

yelp_api = YelpAPI(_KEY)
locations= [i + ', CA' for i in data.Name.to_list()]
_cat = ["dentists"]
radius = 5 * 1610 ## 5 miles

for location in locations:
    _results = yelp_api.search_query(location=location, categories=_cat, radius=radius, offset=0, limit=50)
    
    businesses = _results['businesses']
    total = _results["total"]
    ## print (total, len(businesses))
    
    for page in range(1, int(total / 50) + 1):
        _r = yelp_api.search_query(location=location, categories=_cat, radius=radius, offset=page*50, limit=50)
        ## print (page, _r["total"], len(_r["businesses"]))
        businesses += _r['businesses']
    
    file_name = 'crawled_data/' + slugify(location) + '.json'
    with open(file_name, 'w') as f:
        json.dump(businesses, f, sort_keys=True, indent=4, separators=(',', ': '))
        
  
    print ("Found {} in {}".format(len(businesses), location))
    
```
